```python
img1=cv2.imread("resources/data/fav.png")
img=cv2.cvtColor(img1,cv2.COLOR_BGR2HSV)
figure_size=7
result=cv2.blur(img,(figure_size,figure_size))
plt.figure(figsize=(11,6))
plt.subplot(121),plt.imshow(cv2.cvtColor(img,cv2.COLOR_HSV2RGB)),plt.title('original')
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(cv2.cvtColor(result,cv2.COLOR_HSV2RGB)),plt.title('blurred')
plt.xticks([]),plt.yticks([])
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
img1=cv2.imread("resources/data/fav.png")
(height,width)=img1.shape[0:2]
zoom=cv2.resize(img1,None,fx=1.5,fy=1.5,interpolation=cv2.INTER_CUBIC)
plt.imshow(zoom)
```

```python
zoom=cv2.resize(img1,(int(width*2),int(height*2)),fx=1.5,fy=1.5,interpolation=cv2.INTER_CUBIC)
plt.imshow(zoom)
```

```python
import os
from os import listdir
import cv2
import matplotlib.pyplot as plt
path=r"C:\Users\ASUS\Desktop\img"
images=os.listdir(path)
print('No. of images:',len(images))  #no. of imagesi= in the folder
print(images)
print(type(images))
img_data=[]
for img in images:
    img_arr=cv2.imread(os.path.join(path,img))
    print(img_arr)
    plt.figure()
    plt.imshow(img_arr)  #reading all images 

```

```python
img_data=[]
for img in images:
    img_arr=cv2.imread(os.path.join(path,img))
    img_data.append(img_arr)
for i in range(len(img_data)):
    plt.subplot(6,4,i+1)
    plt.imshow(img_data[i])
```
